```python
wordstring = ' '.join(text)
wordfreq = []
for w in text:
    wordfreq.append(text.count(w))

#print("String\n" + wordstring +"\n")
#print("List\n" + str(text) + "\n")
#print("Frequencies\n" + str(wordfreq) + "\n")
#print("Pairs\n" + str(list(zip(text, wordfreq))))
Frequencies_df = pd.DataFrame(list(zip(text, wordfreq)), columns = ['Text', 'Frequencies'])
Frequencies_df = Frequencies_df.drop_duplicates()
#Frequencies_df = Frequencies_df.set_index('Text')
Frequencies_df.sort_values('Frequencies', ascending=False).head(10)
```
